## Generate artificial data


```python

def build_data():
    """
    # Python yield:
    # for i in range(100): rec= .... yield rec: 
    # generates 100 dicts rec, each dict is a record (row, sample)
    #   for j in range(50): rec[str(j)]=j 
    #   generates additional 50 key value pairs, becoming column names (keys) and data (values)
    # yield rec
    # returns the dict rec after each iteration to the caller list(build_data()), which builds a list of dicts   
    """
    for i in range(100):
        rec = dict(date=pd.Timestamp('20000101'), security_id=i, foo=i, bar=1.2345678)
        for j in range(50):
            rec[str(j)] = j
        yield rec
```

```python
df = pd.DataFrame(list(build_data())).set_index(['date','security_id'])
df.describe(datetime_is_numeric=True)
dtale.show(df)
```

- Data is downloaded and then stored somewhere for subsequent use

```python

import tensorflow_datasets as tfds
tfds.list_builders()
```


### There are several ways to load the data. 
- See:
    https://www.tensorflow.org/datasets/overview
- For tensors use: ds = tfds.load (tensorflow)
- For visualization use: tfds.as_dataframe  (pandas)

```python

ds, info = tfds.load('mnist', split='train', with_info=True)
df_mnist = tfds.as_dataframe(ds.take(150), info)
display(df_mnist)
dtale.show(df_mnist)
```
